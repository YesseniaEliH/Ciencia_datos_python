## Ejercicio Nº 3:
- Crear una función que permita generar 3 variables
cualitativas ordinales a partir de las variables numéricas con
los siguientes criterios:

1. Si es mayor a la “media + 1*desviación estándar”
→ Alto
2. Si está entre “media - 1*desviación estándar” y
“media + 1*desviación estándar” → Medio
3. Si es menor a la “media - 1*desviación estándar”
→ Bajo
Nota: Mostrar con las nuevas variables generadas.
Elegir con libertad las variables a trabajar.
La función debe contar con su respectiva documentación.

```python
def calculo_variable_cualitativa(notas, media, desviacion):
    """
    Parámetros
    ----------
    `notas` : serie(este debe tener una columna con las notas por alumno de un mes específico)
    `media` : serie(este debe tener una columna con las medias de las notas por alumno)
    `desviacion` : serie(este debe tener una columna con la desviacion estandar por alumno)

    Tiene como finalidad generar una serie donde se asigne una variable cualitativa ordinal (alto, medio o bajo) de acuerdo al siguiente cálculo:
    1. Si es mayor a la “media + 1*desviación estándar” → Alto
    2. Si está entre “media - 1*desviación estándar” y “media + 1*desviación estándar” → Medio
    3. Si es menor a la “media - 1*desviación estándar” → Bajo
    """

    resultado = []
    for i in range(len(notas)):
        descripcion = 'Alto' if (notas[i] > (media[i] + 1*desviacion[i])) else 'Medio' if (notas[i] > (media[i] - 1*desviacion[i])) else 'Bajo'
        resultado.append(descripcion)
    return resultado
# Construimos un diccionario con los promedio mensuales de los alumnos de una universidad
dict_promedios_calculo={
    'enero':{'alumno1':14, 'alumno2':17, 'alumno3':15},
    'febrero':{'alumno1':12, 'alumno2':15, 'alumno3':16},
    'marzo':{'alumno1':17, 'alumno2':18, 'alumno3':13},
    'abril':{'alumno1':15, 'alumno2':16, 'alumno3':14},
    'mayo':{'alumno1':13, 'alumno2':17, 'alumno3':15},
    'junio':{'alumno1':16, 'alumno2':16, 'alumno3':16},
    'julio':{'alumno1':12, 'alumno2':17, 'alumno3':14},
    'agosto':{'alumno1':15, 'alumno2':18, 'alumno3':16},
    'setiembre':{'alumno1':13, 'alumno2':13, 'alumno3':12},
    'octubre':{'alumno1':15, 'alumno2':15, 'alumno3':14},
    'noviembre':{'alumno1':16, 'alumno2':16, 'alumno3':14},
    'diciembre':{'alumno1':18, 'alumno2':18, 'alumno3':15}
}
# Generamos el dataframe del diccionario
df= pd.DataFrame(index=dict_promedios_calculo.keys())
df['alumno1'] = [value['alumno1'] for value in dict_promedios_calculo.values()]
df['alumno2'] = [value['alumno2'] for value in dict_promedios_calculo.values()]
df['alumno3'] = [value['alumno3'] for value in dict_promedios_calculo.values()]
df_promedios_calculo = df.T.copy()
df_promedios_calculo
```

```python
# Cálculo de la media
df_promedios_calculo_media = df_promedios_calculo.T.mean(axis=0)
df_promedios_calculo_media
```

```python
# Cálculo de la desviacion estándar
df_promedios_calculo_desviacion = df_promedios_calculo.T.std()
df_promedios_calculo_desviacion
```

```python
# Unimos en un solo dataframe
df_promedios_calculo_total = df.T.copy()
df_promedios_calculo_total['media'] = df_promedios_calculo_media.round(2)
df_promedios_calculo_total['desviacion_estandar'] = df_promedios_calculo_desviacion.round(2)
df_promedios_calculo_total
```

```python
# Creamos una nueva columna con el resultado de la funcion
df_promedios_calculo_total.assign(media_descripcion = calculo_variable_cualitativa(df_promedios_calculo_total['enero'], df_promedios_calculo_total['media'], df_promedios_calculo_total['desviacion_estandar']) )

```

## Ejercicio Nº 4:
Crear una función que muestre los siguientes gráficos
(simultáneamente) para las variables numéricas, según una
variable cualitativa:
- Cajas y bigotes
- Barras de la media
- Histograma con Curva

La función debe contar con su respectiva documentación.

```python
# Construimos un dataframe
df_ventas_gastos= pd.DataFrame()
df_ventas_gastos['Mes'] = ('Enero','Febrero','Marzo','Abril','Mayo')
df_ventas_gastos['Ventas'] = (30500,35600,28300,33900,32500)
df_ventas_gastos['Gastos'] = (22000,26450,20100,25700,22450)
df_ventas_gastos
```

```python
def genera_graficos(x_name = None,y_name = None,df_ = None):
    """
    Parámetros
    ----------
    `x_name` : array
    `y_name` : array
    `df_` : dataframe    

    La finalidad es pedir el ingreso de la columna del eje x y la o las columnas del eje y el dataframe
    El resultado son tres gráficas (cajas y bigotes, barras, histograma con curva)
    """    
    #gráfico cajas y bigotes
    df_.plot.box()
    #gráfico barras
    g =df_[y_name].plot(kind='bar')
    g.set_xlabel(x_name[0])
    #gráfico histograma con curva
    sns.displot(data=df_, x=y_name[0], kde=True)

    
```

```python
genera_graficos(['Mes'], ['Ventas', 'Gastos'],df_ventas_gastos)
```
